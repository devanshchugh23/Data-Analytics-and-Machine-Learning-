# 4. Use Python library mlxtend and convert the transactions into a format that can be used in the Apriori method for finding frequent itemsets.

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

# 5. Find top selling items with minimum support of 2%.

```python
te = TransactionEncoder()
te_ary = te.fit(df[0]).transform(df[0])
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
apriori(df, min_support=0.02,use_colnames = True).sort_values(by = "support" , ascending = False).head()
```

# 6. Find all frequent itemsets with minimum support of 5%.


```python
apriori(df, min_support=0.05,use_colnames = True).sort_values(by = "support")
```

# 7. Find all frequent itemsets of length 2 with minimum support of 2%.

```python
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[frequent_itemsets["length"] == 2]
```

# 8. Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order

```python
var = apriori(df, min_support=0.02, use_colnames=True)
var = association_rules(var, metric="lift",min_threshold = 1)
var.sort_values(by="confidence" , ascending =False).head(10)
```
